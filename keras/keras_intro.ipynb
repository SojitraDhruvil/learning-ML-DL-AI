```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create a sequential model
model = Sequential()
# add a dense layer with 64 units and ReLU activation
layer1 = Dense(32, activation='relu', input_shape=(30,))
layer2 = Dense(16,activation = 'relu' )
output_layer = Dense(1, activation='sigmoid')
model.add(layer1)
model.add(layer2)
model.add(output_layer)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

canser = datasets.load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(canser.data, canser.target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
model.fit(x_train, y_train, epochs=100 , batch_size=50 , validation_data=(x_test, y_test))
```

```python
accuracy = model.evaluate(x_test, y_test)[1]
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
pridictions = model.predict(x_test)
print(pridictions[:5])
```

```python
score = model.evaluate(x_test, y_test)
print(score)
print(f"Test Loss: {score[0]:.4f}, Test Accuracy: {score[1]:.4f}")
```
