```python
import numpy as np
```

```python
X = np.array([[0,0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0,1,1,0]]).T
print(X , "shape of X is" , X.shape)
print(Y, "shape of Y is" , Y.shape)
```

```python

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def derivative_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
    
```

```python
# Weights and bias initialization
wh = 2 * np.random.rand(2, 2) - 1  # Weights for hidden layer
bh = 2 * np.random.rand(1, 2) - 1  # Bias for hidden layer (one bias per hidden neuron)
wo = 2 * np.random.rand(2, 1) - 1  # Weights for output layer
bo = 2 * np.random.rand(1, 1) - 1  # Bias for output layer

lerning_rate = 0.1
```

```python
# Updated Training Loop
for iteration in range(100000):
    # Forward Propagation
    input_layer = X
    hidden_layer_input = np.dot(input_layer, wh) + bh
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, wo) + bo
    predicted_output = sigmoid(output_layer_input)

    # Backpropagation - Output Layer
    # Error: (Predicted - Actual)
    error_output_layer = predicted_output - Y
    # Gradient: Error * derivative of sigmoid at the output
    d_output = error_output_layer * (predicted_output * (1 - predicted_output))

    # Backpropagation - Hidden Layer
    # Error: Contribution of hidden neurons to output error
    error_hidden_layer = np.dot(d_output, wo.T)
    # Gradient: Error * derivative of sigmoid at the hidden layer
    d_hidden = error_hidden_layer * (hidden_layer_output * (1 - hidden_layer_output))

    # Weight and Bias Updates
    # FIX: Dot product of hidden layer output and the output gradient
    changes_output_layer = np.dot(hidden_layer_output.T, d_output)
    changes_output_bias = np.sum(d_output, axis=0, keepdims=True)

    changes_hidden_layer = np.dot(input_layer.T, d_hidden)
    changes_hidden_bias = np.sum(d_hidden, axis=0, keepdims=True)

    # Gradient Descent update
    wo = wo - lerning_rate * changes_output_layer
    bo = bo - lerning_rate * changes_output_bias
    wh = wh - lerning_rate * changes_hidden_layer
    bh = bh - lerning_rate * changes_hidden_bias
```

```python
input_layer = X
hidden_layer_input = np.dot(input_layer, wh) + bh
hidden_layer_output = sigmoid(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, wo) + bo
predicted_output = sigmoid(output_layer_input)


print("wh wb wo bo ")
print(wh)
print(bh)
print(wo)
print(bo)

print("Predicted output after training:")
print(predicted_output)
```
